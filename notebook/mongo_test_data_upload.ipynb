### Connect to mongodb

```python
import yaml
from urllib.parse import quote_plus
#quote plus is used to encode special characters in username and password, 
#if username and password contains special characters they are escaped so that there is no contradiction
#with ? and @ used in mongodd connection string

config = yaml.safe_load(open(r"../project_cred.yaml"))

user = quote_plus(config["mongo"]["user"])
password = quote_plus(config["mongo"]["password"])
host = config["mongo"]["host"]
cluster = config["mongo"]["cluster"]

CONNECTION_URL = f"mongodb+srv://{user}:{password}@{host}/?appName={cluster}"
DB_NAME = "Proj1"
COLLECTION_NAME = "Proj1-Data"

```

```python
data = df.to_dict(orient="records")
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```
